3.จำนวนผู้ที่แต่งงานแล้ว,ยังไม่แต่งงานและระบุไม่ได้

```python
# นับจำนวนผู้ที่แต่งงานแล้วและยังไม่แต่งงานโดยใช้ข้อมูล Name
married_counts = 0
single_counts = 0
unspecified_counts = 0

for index, row in df.iterrows():
    name = row['Name']

    # ตรวจสอบคำคุณศัพท์ในชื่อเพื่อระบุสถานะการแต่งงาน
    if 'Mrs' in name:
        married_counts += 1
    elif 'Miss' in name or 'Master' in name:
        single_counts += 1
    else:
        unspecified_counts += 1
print("จำนวนผู้แต่งงานแล้ว:", married_counts)
print("จำนวนผู้ยังไม่แต่งงาน:", single_counts)
print("จำนวนผู้ระบุไม่ได้:", unspecified_counts)
marital_status_counts = [married_counts, single_counts, unspecified_counts]
```

```python
# Bars
width = 0.35  # the width of the bars
situations = ['Married', 'Single', ' Unspecified']
data = [1, 2, 3]  # the bar positions on x-axis

plt.bar(data, marital_status_counts, width, edgecolor='black',
        color=['lavender', 'pink', 'silver'])
plt.xticks(data, situations)

for p in plt.gca().patches:
    plt.gca().annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')
plt.show()
```

```python
# สร้างฟังก์ชันเพื่อแปลงราคาตั๋วในระบบหลักหน่วย, หลักสิบ, และหลักร้อย
def format_fare(fare):
    return f"${fare:.0f} "


# กรองข้อมูลเฉพาะผู้โดยสารที่มีห้องโดยสาร
passengers_with_cabin = df[df['Cabin'].notna()]

# คำนวณราคาตั๋วเฉลี่ยในระบบหลักหน่วย, หลักสิบ, และหลักร้อย
average_fare = passengers_with_cabin['Fare'].mean()
formatted_fare = format_fare(average_fare)

# แสดงผลราคาตั๋วเฉลี่ย
print(f"ราคาตั๋วเฉลี่ยสำหรับผู้โดยสารที่มีห้องโดยสาร: {formatted_fare}")
```

```python
import pandas as pd

# สร้าง DataFrame ดั่งตัวอย่าง
data = {'Pclass': [1, 2, 3, 1, 2, 3, 1, 2, 3],
        'Age': [35, 22, 18, 40, 28, 21, 38, 50, None],
        'Survived': [1, 1, 0, 0, 1, 0, 1, 0, 1]}
df = pd.DataFrame(data)

# แปลงคอลัมน์ 'Pclass' เป็น category และเพิ่มหมวดหมู่ 'ระบุไม่ได้' กรณีที่ข้อมูลหาย
df['Pclass'] = pd.Categorical(df['Pclass'], categories=[1, 2, 3, 'ระบุไม่ได้'])

# นับจำนวนผู้รอดชีวิตและผู้เสียชีวิตตามชั้นและช่วงอายุ
survival_by_class_age = df.groupby(['Pclass', pd.cut(df['Age'], [0, 17, 65, pd.np.inf], right=False)])[
    'Survived'].value_counts().unstack().fillna(0)

# แสดงผลลัพธ์
print("จำนวนผู้รอดชีวิตและผู้เสียชีวิตตามชั้นและช่วงอายุ:")
print(survival_by_class_age)
```

90.จำนวนผู้รอดชีวิต และผู้เสียชีวิต ในแต่ละชั้นเป็นเพศหญิงหรือชาย


```python
import pandas as pd

# สร้าง DataFrame สำหรับผู้รอดชีวิตและผู้เสียชีวิตตามชั้นและการมีห้องโดยสาร
survival_by_class_with_cabin = pd.DataFrame(
    columns=['Pclass', 'จำนวนรอด', 'จำนวนเสีย'])
survival_by_class_without_cabin = pd.DataFrame(
    columns=['Pclass', 'จำนวนรอด', 'จำนวนเสีย'])

for pclass in range(1, 4):
    with_cabin_df = df[(df['Pclass'] == pclass) & (df['Cabin'].isna())]
    without_cabin_df = df[(df['Pclass'] == pclass) & (df['Cabin'].isna())]

    survived_with_cabin = with_cabin_df[with_cabin_df['Survived'] == 1].shape[0]
    not_survived_with_cabin = with_cabin_df[with_cabin_df['Survived'] == 0].shape[0]

    survived_without_cabin = without_cabin_df[without_cabin_df['Survived'] == 1].shape[0]
    not_survived_without_cabin = without_cabin_df[without_cabin_df['Survived'] == 0].shape[0]

    survival_by_class_with_cabin = survival_by_class_with_cabin.append({
        'Pclass': pclass,
        'จำนวนรอด': survived_with_cabin,
        'จำนวนเสีย': not_survived_with_cabin
    }, ignore_index=True)

    survival_by_class_without_cabin = survival_by_class_without_cabin.append({
        'Pclass': pclass,
        'จำนวนรอด': survived_without_cabin,
        'จำนวนเสีย': not_survived_without_cabin
    }, ignore_index=True)

# แสดงผลลัพธ์
print("จำนวนผู้รอดชีวิตและผู้เสียชีวิตตามชั้นที่มีห้องโดยสารและไม่มีห้องโดยสาร:")
print("โดยมีห้องโดยสาร:")
print(survival_by_class_with_cabin)
print("\nโดยไม่มีห้องโดยสาร:")
print(survival_by_class_without_cabin)
```
